# Etapa de preparação e processamento dos dados

```python
import pandas as pd
```

```python
# lendo os arquivos csv
df1 = pd.read_csv('dados/dados-brasileirao.csv')
df2 = pd.read_csv('dados/brasileirao_odds.csv')
```

```python
df1.head()
```

```python
# pegando dados somente de 2021, 2022, 2023 de df1 
df_cbf = df1[df1['temporada'].isin([2021, 2022, 2023])].reset_index(drop=True)


df_cbf.head()
```

```python
# contando o numero de linhas para a temporada 2021
df_cbf[df_cbf['temporada'] == 2023].shape[0]
```

```python
# contando o numero de linhas para a temporada 2022
df_cbf[df_cbf['temporada'] == 2022].shape[0]
```

```python
# contando o numero de linhas para a temporada 2023
df_cbf[df_cbf['temporada'] == 2023].shape[0]
```

```python
# Ajustar os nomes das colunas de df2 para corresponderem aos de df_cbf para a fusão
df2.rename(columns={'Year': 'ano', 'Time': 'hora', 'Home Team': 'mandante', 
                    'Away Team': 'visitante', 'Home Score': 'gol_mandante', 
                    'Away Score': 'gol_visitante'}, inplace=True)

df2.head()
```

```python
# listando os times únicos de df1
times = df_cbf['mandante'].unique()

print(times)
```

```python
# listando os times únicos de df2
times_odds = df2['mandante'].unique()

print(times_odds)
```

```python
# Dicionário de mapeamento baseado nas listas fornecidas
name_mapping = {
    'America MG': 'América Fc Saf - MG',
    'Atletico GO': 'Atlético - GO',
    'Bragantino': 'Red Bull Bragantino - SP',
    'Juventude': 'Juventude - RS',
    'Fluminense': 'Fluminense - RJ',
    'Fortaleza': 'Fortaleza - CE',
    'Gremio': 'Grêmio - RS',
    'Palmeiras': 'Palmeiras - SP',
    'Santos': 'Santos - SP',
    'Sport Recife': 'Sport - PE',
    'Chapecoense-SC': 'Chapecoense - SC',
    'Cuiaba': 'Cuiabá Saf - MT',
    'Flamengo RJ': 'Flamengo - RJ',
    'Internacional': 'Internacional - RS',
    'Athletico-PR': 'Athletico Paranaense - PR',
    'Sao Paulo': 'São Paulo - SP',
    'Ceara': 'Ceará - CE',
    'Atletico-MG': 'Atlético Mineiro Saf - MG',
    'Bahia': 'Bahia - BA',
    'Corinthians': 'Corinthians - SP',
    'Goias': 'Goiás - GO',
    'Botafogo RJ': 'Botafogo - RJ',
    'Avai': 'Avaí - SC',
    'Coritiba': 'Coritiba - PR',
    'Cruzeiro': 'Cruzeiro Saf - MG',
    'Vasco': 'Vasco da Gama S.a.f. - RJ'
}

# Aplicar o mapeamento para as colunas 'mandante' e 'visitante' no df2
df2['mandante'] = df2['mandante'].map(name_mapping)
df2['visitante'] = df2['visitante'].map(name_mapping)

# Verifique se há alguma entrada que não foi mapeada corretamente
print(df2[df2['mandante'].isnull() | df2['visitante'].isnull()])

df2.head()
```

```python
# contando o numero de linhas para a temporada 2022
df2[df2['ano'] == 2023].shape[0]
```

```python
df_merged = pd.merge(df_cbf, df2, on=['ano', 'mandante', 'visitante'], how='right')

df_merged.head()
```

```python
# Remover linhas onde a coluna 'serie' é nula
df_merged = df_merged[df_merged['serie'].notna()]


# contando o numero de linhas do dataframe df_merged
df_merged.shape[0]
```

```python
# contando numero de linhas para o ano 2021
print(df_merged[df_merged['ano'] == 2021].shape[0])
```

```python
# contando numero de linhas para o ano 2022
print(df_merged[df_merged['ano'] == 2022].shape[0])
```

```python
# contando numero de linhas para o ano 2023
print(df_merged[df_merged['ano'] == 2023].shape[0])
```

```python
# salvando o dataframe em um arquivo csv
df_merged.to_csv('dados/brasileirao_2021_2023.csv', index=False)

#  salvando o dataframe em um arquivo excel
df_merged.to_excel('dados/brasileirao_2021_2023.xlsx', index=False)
```

```python
import pandas as pd
import json

# Carregar o arquivo JSON
with open('dados/aa.json', 'r') as file:
    json_data = json.load(file)

# Criar uma lista para armazenar dados de cada jogo
game_list = []

# Iterar sobre cada jogo no JSON
for game in json_data['d']['rows']:
    game_info = {
        'Game ID': game['id'],
        'Home Team': game['home-name'],
        'Away Team': game['away-name'],
        'Home Score': game['homeResult'],
        'Away Score': game['awayResult'],
        'Result': game['result'],
        'Odds 1': game['odds'][0]['avgOdds'] if game['odds'] else None,
        'Odds X': game['odds'][1]['avgOdds'] if len(game['odds']) > 1 else None,
        'Odds 2': game['odds'][2]['avgOdds'] if len(game['odds']) > 2 else None
    }
    game_list.append(game_info)

# Criar DataFrame a partir da lista de jogos
df_games = pd.DataFrame(game_list)

# Exibir o DataFrame
print(df_games.head())
```
